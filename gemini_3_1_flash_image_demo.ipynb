```python
!pip install google-genai Pillow
from google.colab import auth
auth.authenticate_user()

from google import genai
from PIL import Image
from io import BytesIO
import os

project_id = 'YOUR_PROJECT_ID'  # @param {type:"string"}
client = genai.Client(vertexai=True, project=project_id, location="us-central1")
```

### 2. Generate Image with Thinking Levels

Gemini 3.1 allows you to see the model's reasoning before it generates the final image. We set `thinking_level='HIGH'` for complex artistic prompts.

```python
prompt = "A photorealistic image of a futuristic sustainable city with vertical forests and flying electric vehicles, 4K, cinematic lighting."

print(f"Generating image with Thinking Level: HIGH...")

try:
    response = client.models.generate_content(
        model="gemini-3.1-flash-image",
        contents=[prompt],
        config=genai.types.GenerateContentConfig(
            thinking_level="HIGH" # New feature: High reasoning for better prompt adherence
        )
    )

    # Display the Thinking Process (if available in the response)
    if hasattr(response, 'thought'):
        print("--- Model Thinking Process ---")
        print(response.thought)
        print("------------------------------")

    # Extract and display image bytes
    for part in response.candidates[0].content.parts:
        if part.inline_data:
            img = Image.open(BytesIO(part.inline_data.data))
            display(img)
            print("Success: Image generated and displayed.")
except Exception as e:
    print(f"Error: {str(e)}")
    print("[Note: Ensure Gemini 3.1 Flash Image is enabled in your project's Vertex AI Model Garden]")
```

### 3. Things to remember or know
- **Thinking Process**: Setting `thinking_level` to `HIGH` improves the composition and detail of generated images by forcing the model to 'plan' the output.
- **SDK Migration**: Use the `google-genai` package for all Gemini 3.x features to avoid deprecation warnings from older libraries.
- **Native Multimodality**: Gemini 3.1 models handle image generation, editing, and understanding in a single unified architecture.
