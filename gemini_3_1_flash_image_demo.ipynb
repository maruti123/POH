## Use Case
A marketing team needs to generate localized promotional images. We'll show how to generate and display a high-resolution futuristic city image using the flagship 3.1 Pro model.

```python
!pip install google-genai Pillow
from google.colab import auth
auth.authenticate_user()

from google import genai
from PIL import Image
from io import BytesIO
import os

project_id = 'YOUR_PROJECT_ID'  # @param {type:"string"}
client = genai.Client(vertexai=True, project=project_id, location="us-central1")
```

### 2. Generate and Display Image

Gemini 3.1 Pro handles generation and understanding in a single unified call.

```python
prompt = "A photorealistic image of a futuristic sustainable city with vertical forests and flying electric vehicles, 4K, cinematic lighting."

print(f"Generating image with Gemini 3.1 Pro...")

try:
    response = client.models.generate_content(
        model="gemini-3.1-pro",
        contents=[prompt]
    )

    # Extract and display image bytes from the response candidates
    for part in response.candidates[0].content.parts:
        if part.inline_data:
            img = Image.open(BytesIO(part.inline_data.data))
            display(img)
            print("Success: Image generated and displayed.")
except Exception as e:
    print(f"Error: {str(e)}")
    print("[Note: Ensure Gemini 3.1 Pro is enabled in your project's Vertex AI Model Garden]")
```
