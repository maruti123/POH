## Use Case
A marketing team needs to generate localized promotional images for a campaign quickly and at scale. We'll generate an image of a 'Modern sustainable city in 2026'.

```python
from google.colab import auth
auth.authenticate_user()

import vertexai
from vertexai.generative_models import GenerativeModel, Part
from IPython.display import display, Image

project_id = 'YOUR_PROJECT_ID'  # @param {type:"string"}
vertexai.init(project=project_id, location="us-central1")
```

### 2. Generate and Display Image with Gemini 3.1 Flash Image

We use the `generate_content` method which returns the image bytes directly in the `Part` object.

```python
model = GenerativeModel("gemini-3.1-flash-image")

prompt = "A high-resolution, photorealistic image of a futuristic sustainable city with vertical forests, flying electric vehicles, and clear blue skies, in 4K resolution."

print(f"Generating image for prompt: {prompt}...")

try:
    response = model.generate_content(prompt)
    
    # Extract image bytes from the response parts
    for part in response.candidates[0].content.parts:
        if part.inline_data:
            display(Image(data=part.inline_data.data))
            print("Success: Image generated and displayed.")
except Exception as e:
    print(f"Error generating image: {str(e)}")
    print("[Note: In a live environment, ensure the project has the Gemini 3.1 Flash Image model enabled in Vertex AI Model Garden]")
```
