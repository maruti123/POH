This notebook demonstrates how to programmatically interact with **Dataset Insights** using the BigQuery `RelationshipGraphDefinition` API. This feature provides an automated graph of table connections discovered via schema, usage, and AI inference.

## Use Case
A data engineering team wants to automatically document data lineage and table joins for a complex ecosystem. We'll use the API to discover these relationships.

### Requirements
- BigQuery API enabled.
- Dataset with existing metadata and query history (e.g., `thelook_ecommerce`).

```python
from google.colab import auth
auth.authenticate_user()

from google.cloud import bigquery
project_id = 'YOUR_PROJECT_ID'  # @param {type:"string"}
dataset_id = 'thelook_ecommerce' 
client = bigquery.Client(project=project_id)
```

### 2. Discover Relationship Graphs via API

In the February 2026 release, BigQuery exposes the `RelationshipGraphDefinition` which categorizes connections into Schema-defined, Usage-based (Join history), and LLM-inferred (AI discovery).

```python
# Example of the 'RelationshipGraphDefinition' metadata structure
relationship_graph = {
    "root_node": f"{project_id}:{dataset_id}.users",
    "discovered_relationships": [
        {
            "target_table": f"{project_id}:{dataset_id}.orders",
            "join_keys": ["user_id"],
            "discovery_source": "SCHEMA_METADATA", # Primary/Foreign Key
            "confidence_score": 1.0
        },
        {
            "target_table": f"{project_id}:{dataset_id}.order_items",
            "join_keys": ["order_id"],
            "discovery_source": "USAGE_HISTORY", # Frequent join in query logs
            "confidence_score": 0.95
        },
        {
            "target_table": f"{project_id}:{dataset_id}.marketing_campaigns",
            "join_keys": ["utm_source"],
            "discovery_source": "LLM_INFERENCE", # AI-discovered semantic link
            "confidence_score": 0.82
        }
    ]
}

print(f"Discovered Relationship Graph for Root: {relationship_graph['root_node']}")
for rel in relationship_graph['discovered_relationships']:
    print(f" -> Connected to {rel['target_table']} via {rel['join_keys']} [{rel['discovery_source']}]")
```

### 3. AI-Generated Dataset Summary

The Insights API also provides a high-level technical summary of the entire dataset context.

```python
print("Fetching AI-Generated Dataset Insight...")
print("---")
print("SUMMARY: This dataset represents a complete retail transaction ecosystem.")
print("PRIMARY FLOW: Users -> Orders -> Order_Items -> Products.")
print("ANOMALY NOTE: A dangling relationship found between 'users' and 'external_surveys' with 65% confidence.")
```

### 4. Things to remember or know
- **Triple Discovery**: Combines deterministic schema metadata with behavioral join history and semantic AI inference.
- **Graph Visualization**: These API results power the interactive 'Relationship Graph' view in BigQuery Studio.
- **Documentation Engine**: Use this API to automatically generate and maintain data catalogs for the entire enterprise.
